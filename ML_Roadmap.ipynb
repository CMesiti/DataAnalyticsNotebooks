```python
# Install dependencies as needed:
# pip install kagglehub[pandas-datasets]
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "sentimentdataset.csv"

# Load the latest version
df_media = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "eshummalik/socialbuzz-sentiment-analytics",
  file_path,
  # Provide any additional arguments like
  # sql_query or pandas_kwargs. See the
  # documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)

print("First 5 records:")
df_media.head()
```

```python
# df_media = df_media.drop(columns=['hashtag_count', 'index'])

df_media.isnull().sum() #Counts only values that are 'TRUE' meaning +1 for each NULL

df_media.count() #Counts all values expect missing values. Essentially the inverse of sum for this usecase.

df_media.columns = df_media.columns.str.lower() #The str accessor is a vectorization

df_media[['likes', 'retweets']].describe() #descibe is the best way to get full summary of numerical data
```

```python
from sklearn.linear_model import LinearRegression

#lets see if there is a correlation between likes and retweets.
#strong positive correlation
X = df_media[['likes']] #keeps this as a dataframe since the X input needs to be 2 dimensions.
y = df_media['retweets']

print(X.shape)
reg = LinearRegression().fit(X, y)
print(reg.score(X, y))
print(reg.coef_)

sns.regplot(data=df_media, x="likes", y="retweets")
```

```python
#lets see how our model reacts with general numbers
print(reg.feature_names_in_)
x_data = np.linspace(X.min(), X.max(), 100).squeeze()
x_data = pd.DataFrame({reg.feature_names_in_[0]:x_data})
y_pred = reg.predict(x_data)

sns.scatterplot(df_media, x=df_media['likes'], y=y)
plt.plot(x_data, y_pred)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
#first lets check if developers focus on a specific genre
#We will do this by counting the genres for each dev.
mod_df = df.copy()
mod_df.groupby('Developers')['Genres'].value_counts().reset_index().sort_values(by='count', ascending=False)
mod_df = mod_df.dropna(subset=['Genres'])
mod_df = mod_df.dropna(subset=['About the game'])
mod_df = mod_df.dropna(subset=['Categories'])

X = mod_df[['Price', 'Required age', 'About the game', 'Categories']]
y = mod_df['Genres']
xtrain, xtest, ytrain, ytest = train_test_split(X, y, train_size=.80, test_size = .20)
```
